## LIBRARY YANG DIBUTUHKAN

```python
import pandas as pd
```

## MEMBACA DAN MENAMPILKAN DATASET

```python
df = pd.read_csv('movie_sample_dataset.csv')

df
```

## MENGECEK TIPE DATA DAN MISSING VALUE

```python
df.info()
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

## CLEANING DATA

```python
# df['color'] = df['color'].str.capitalize()
df.loc[:, 'color'] = df['color'].str.strip().str.capitalize()


df = df[(df['duration'] >= 0) & 
        (df['gross'] >= 0) & 
        (df['title_year'] >= 0) & 
        (df['budget'] >= 0) & 
        (df['imdb_score'] >= 0) & 
        (df['movie_facebook_likes'] >= 0)]

df = df.replace('N/A', pd.NA)

# df.dropna(subset=['duration', 'gross', 'title_year', 'budget', 'imdb_score', 'movie_facebook_likes'], inplace=True)
df.dropna(inplace=True)
df
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

## TRANSFORM DATA

```python
df.info()
```

```python
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
df['gross'] = pd.to_numeric(df['gross'], errors='coerce')
print(df.dtypes)
```

```python
df['actors'] = df['actors'].str.replace(',', ', ', regex=False)

df['genres'] = df['genres'].str.replace('|', ', ', regex=False)

df
```

## FINAL DATASET

```python
df.to_csv('movie_dataset_cleaned.csv', index=False)

df
```
